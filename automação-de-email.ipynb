# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Passo a passo p/ resolver desafio:

Passo 1: Entrar no sistema da empresa
 https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing
 
Passo 2: Navegar ao local onde está a base de dados

Passo 3: Baixar planilha de vendas

Passo 4: Calcular faturamento e quantidade de produtos vendidos da base

Passo 5: Enviar email p/ diretoria com os dados que calculamos

```python
import pyautogui
import time
import pyperclip

pyautogui.PAUSE=1
# Passo 1
# Abrir o Opera
pyautogui.hotkey('win')
pyautogui.write('Opera')
pyautogui.press('enter')
# Entrar no link do sistema
link= 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'
pyperclip.copy(link)
pyautogui.hotkey('ctrl','v')
pyautogui.press('enter')
# Passo 2
time.sleep(5)
pyautogui.click(416, 305, clicks=2)
# Passo 3
time.sleep(2)
pyautogui.click(427, 385) #clicar no arquivo
time.sleep(2)
pyautogui.click(1233, 185) #clicar nos 3 pontos
time.sleep(3)
pyautogui.click(983, 592) #clicar no fazer download
time.sleep(10)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
#Passo 4
import pandas as pd

tabela=pd.read_excel(r"C:\Users\ernes\Downloads\Vendas - Dez.xlsx")
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
display(tabela)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5
#Abrir uma página
pyautogui.hotkey('win')
pyautogui.write('Opera')
pyautogui.press('enter')
#Entrar no Gmail
link = "mail.google.com/"
pyperclip.copy(link)
pyautogui.hotkey('ctrl','v')
pyautogui.press('enter')
time.sleep(7)
#Clicar no botão escrever
pyautogui.click(121, 205)
#digitar p/ quem enviar
time.sleep(3)
pyautogui.write('ernestosrf98+diretoria@gmail.com')
pyautogui.press('tab') #escolher email
pyautogui.press('tab') #passar p/ campo de assunto
#digitar o assunto
assunto= 'Relatório de Vendas'
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl','v')
pyautogui.press('tab') #passar p/ campo de escrever email
#digitar o corpo do email
texto_email= f"""
Prezados, bom dia
O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos vendidos foi de: {quantidade:,}
Abs
Ernesto Reis
"""
pyperclip.copy(texto_email)
pyautogui.hotkey('ctrl','v')
#clicar em enviar
pyautogui.hotkey('ctrl','enter')
```
